```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('train.csv',usecols=['X0','X1','X2','X3','X4','X5'])
data.head()
```

```python
# to check the how many unique labels have each variable
for col in data.columns:
    print(col, ':', len(data[col].unique()),'labels')
```

```python
# let's check how many columns we will obtain after one hot encoding of these variable
pd.get_dummies(data,drop_first=True).shape
```

```python
data.X1.value_counts().sort_values(ascending=False).head(20)
```

```python
# let's makes a list with most frequent categories of the variable

top_10=[x for x in data.X1.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
# now make the 10 binary variable
for label in top_10:
    data[label]=np.where(data['X1']==label,1,0)
data[['X1']+top_10].head(40)
```

```python
# get whole set of dummy variables ,for all the catogorical variables
def one_hot_top_x(df,variable, top_x_lebel):
    for lebel in top_x_lebel:
        df[variable+'_'+lebel]=np.where(df[variable]==lebel,1,0)
data=pd.read_csv('train.csv',usecols=['X0','X1','X2','X3','X4','X5'])    
one_hot_top_x(data,'X2',top_10)
data.head()
```
